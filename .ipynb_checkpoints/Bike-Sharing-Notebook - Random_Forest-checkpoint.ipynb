# Encoding categorical variables.

```python
df["jul_sep_jun_aug"] = df["month"].apply(lambda month : 1 if month in ["July","August","September"] else 0)
df["oct_may"] = df["month"].apply(lambda month : 1 if month in ["October","May"] else 0)
df["dec_mar_nov_apr"] = df["month"].apply(lambda month : 1 if month in ["December","March","November","April"] else 0)
df["jan_feb"] = df["month"].apply(lambda month : 1 if month in ["January","February"] else 0)
```

```python
# dropping 'month', 'year' columns as it is redundant now
df.drop(['month'], axis=1, inplace=True)
df.drop(['year'], axis=1, inplace=True)
```

```python
df.head()
```

```python
def encoding_fn(df,feature):
    tdf = pd.get_dummies(df[feature], drop_first=True)
    df.drop(feature, axis=1, inplace=True)
    return pd.concat([df,tdf], axis=1)

df = encoding_fn(df,'season')
df = encoding_fn(df,'weekday')
df = encoding_fn(df,'weathersit')
```

From the above plots we observe that there is outliers in casual, windspeed and hum columns. Let's handle it.

```python
df = df[~((df.windspeed > np.nanpercentile(df.windspeed, 99)) | (df.casual > np.nanpercentile(df['casual'], 99)))]
print(df.shape)
```

```python
sns.boxplot(df["windspeed"])
```

Here we are able to remove a few outliers.

We considered 99prcentile for outlier removal. If we remove more it may affect the information we get from the whole dataset.

```python
X = df.drop(["instant","cnt","casual","registered"], axis=1)
y = df[["cnt"]]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

Even though the numerical variables in our dataset was Minmax scaled, as we have done some outlier removal we perform Minmax scaling.

```python
nums = ['temp','hum', 'windspeed',"atemp"]
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
X_train[nums] = scalar.fit_transform(X_train[nums])
X_test[nums] = scalar.transform(X_test[nums])
```

# FEATURE SELECTION

As our dataset contains both categorical and numeric features, and the target we have to predict is numeric, ANOVA test is the best feature selection method we can follow.

Here we check how each feature is imporatnt for the target prediction using f-scores and p-value.
We will select those feature with high f-score and we will drop those features of which the p-value is greater than 5.

For that we are using SelectKbest with scorefunction f-regression. We are using f-regression as our scoring function is beacause the target is numeric.
Else we will be using f-classif.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
feat_sel = SelectKBest(score_func=f_regression, k=(len(X_train.columns)))
fs = feat_sel.fit(X_train,y_train)
features = pd.DataFrame(X_train.columns)
f_scores = pd.DataFrame(np.round(fs.scores_,4))
p_value = pd.DataFrame(np.round(fs.pvalues_,4))
top_features = pd.concat([features,f_scores,p_value], axis=1)
top_features.columns = ["features","f_scores","p_value"]
top_features
```

```python
top_features[top_features["p_value"]<0.05] , top_features[top_features["p_value"]<0.05].shape
```

```python
best_features = top_features[top_features["p_value"]<0.05]["features"]
best_features
```

```python
best_features.to_csv("Best_Features")
```

Let's build a model with selected important features.

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train[best_features],y_train)
print(rf.score(X_train[best_features],y_train))
```

```python
print(rf.score(X_test[best_features],y_test))
```

```python
y_tr_pred = rf.predict(X_train[best_features])
y_ts_pred = rf.predict(X_test[best_features])
```

```python
plt.scatter(y_train,y_tr_pred)
```

```python
plt.scatter(y_test,y_ts_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("MAE :",mean_absolute_error(y_test,y_ts_pred))
print("MSE :",mean_squared_error(y_test,y_ts_pred))
print("R2 score :",r2_score(y_test,y_ts_pred))
```

Lets try Hyperparameter tuning for better score while predicting

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
param_grid = {'criterion':["squared_error", "mse", "absolute_error", "poisson"],
              'max_depth': [10,20,40],
              'max_features': ['auto', 'sqrt'],
              'min_samples_leaf': [1, 2, 4],
              'min_samples_split': [2,5,8,10],
              'n_estimators': [100,150,200,300]}
              
from sklearn.model_selection import GridSearchCV
gcv = GridSearchCV(estimator=rf, param_grid=param_grid,n_jobs=-1,cv=3,return_train_score=True,verbose=8)
gcv.fit(X_train[best_features], y_train)
```

```python
gcv.best_estimator_
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

# Final Model

```python
y_pred_fin = gcv.predict(X_test[best_features])
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2Score",r2_score(y_test, y_pred_fin))
print("RMSE on test",np.sqrt(mean_squared_error(y_test,y_pred_fin)))
print("MAE :",mean_absolute_error(y_test,y_pred_fin))
```
