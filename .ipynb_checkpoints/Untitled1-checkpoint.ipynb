```python
import pandas as pd
import numpy as np
df = pd.read_csv("day.csv")
df.head()
```

```python
X= df.iloc[:,2:-1]
y=df["cnt"]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
top_features = SelectKBest(score_func=chi2,k=13).fit(X,y)
```

```python
top_features.scores_
```

```python
df_scores = pd.DataFrame(top_features.scores_, columns=["Scores"], index=[X.columns])
```

```python
df_scores.nlargest(13,"Scores")
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
ranked_feature = pd.DataFrame(model.feature_importances_ , columns=["Importance"], index=X.columns)
```

```python
ranked_feature.nlargest(13,"Importance")
```

```python
ranked_feature.nlargest(13,"Importance").plot()
```
