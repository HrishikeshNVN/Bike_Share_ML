```python
import pandas as pd
import numpy as np
import datetime
import pickle as pkl
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("day.csv")
columns = df.columns
shapes = df.shape
infos = df.info()
cat_cols = ['dteday','season','yr','mnth','holiday','weekday','workingday','weathersit']
num_cols = ['temp','atemp','hum','windspeed']

wether_dict = {1:"Clear/Fewclouds", 2:"Mist/Cloudy" , 3:"LightSnow/LightRain", 4:"HeavyRain"}
season_dict = {1:"winter", 2:"spring", 3:"summer", 4:"fall"}
weekday_dict = {0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"}

model_n_params_dict = {
    "Linear_regression":{"model":LinearRegression(), "params":{'n_jobs': [-1],
'fit_intercept': [True,False],
 'normalize': [False,True],
 'positive': [True,False]}},
                                                              
    "Decision_tree_regressor":{"model":DecisionTreeRegressor(),"params":{"criterion": ["mse", "mae"],
              "min_samples_split": [10, 20, 40],
              "max_depth": [2, 6, 8],
              "min_samples_leaf": [20, 40, 100],
              "max_leaf_nodes": [5, 20, 100],
              }},
    "Extra_tree_regressor":{"model":ExtraTreesRegressor(),"params":{'bootstrap': [True,False],
 'criterion': ["mse", "mae"],
 'max_depth': [10,20,40,60,80,100,None],
 'max_features': ['auto','sqrt'],
 'min_samples_leaf': [1,2,4],
 'min_samples_split': [2,5,10],
 'n_estimators': [10,40,50,100],
 'n_jobs': [-1],
 'verbose': [10]}},
    "Random_forest_regressor":{"model":RandomForestRegressor(),"params":{'bootstrap': [True,False],
 'criterion': ["mse", "mae"],
 'max_depth': [10,20,40,60,80,100,None],
 'max_features': ['auto','sqrt'],
 'min_samples_leaf': [1,2,4],
 'min_samples_split': [2,5,10],
 'n_estimators': [10,40,50,100],
 'n_jobs': [-1],
 'verbose': [10]
}}}

def encode_categorical_features(df):

    df["date"] = pd.DatetimeIndex(df["dteday"])
    df["day"] = pd.DatetimeIndex(df["date"]).day
    df["month"] = pd.DatetimeIndex(df["date"]).month_name()
    df["year"] = pd.DatetimeIndex(df["date"]).year

    df["weathersit"].replace(wether_dict, inplace=True)
    df["season"].replace(season_dict, inplace=True)
    df["weekday"].replace(weekday_dict, inplace=True)

    df["jul_sep_jun_aug"] = df["month"].apply(lambda month : 1 if month in ["July","August","September"] else 0)
    df["oct_may"] = df["month"].apply(lambda month : 1 if month in ["October","May"] else 0)
    df["dec_mar_nov_apr"] = df["month"].apply(lambda month : 1 if month in ["December","March","November","April"] else 0)
    df["jan_feb"] = df["month"].apply(lambda month : 1 if month in ["January","February"] else 0)

    df = encoding_fn(df,'season')
    df = encoding_fn(df,'weekday')
    df = encoding_fn(df,'weathersit')
    return df



def remove_no_use_cols(df):
    df.drop(labels=["dteday","mnth","date","month","year","instant","casual","registered"], axis=1, inplace=True)
    return df


def encoding_fn(df,feature):
    tdf = pd.get_dummies(df[feature], drop_first=True)
    df.drop(feature, axis=1, inplace=True)
    return pd.concat([df,tdf], axis=1)

def dep_indep_seperator(df):
    y = df[["cnt"]]
    X = df.drop(["cnt"], axis=1)
    return X,y

def splits_to_train_test(X,y):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
    return X_train,X_test,y_train,y_test


def scale(X_train,X_test):
    scalar = MinMaxScaler()
    X_train[num_cols] = scalar.fit_transform(X_train[num_cols])
    X_test[num_cols] = scalar.transform(X_test[num_cols])
    return X_train, X_test 


def feature_selection(X_train,y_train):
    feat_sel = SelectKBest(score_func=f_regression, k=(len(X_train.columns)))
    fs = feat_sel.fit(X_train,y_train)
    features = pd.DataFrame(X_train.columns)
    f_scores = pd.DataFrame(np.round(fs.scores_,4))
    p_value = pd.DataFrame(np.round(fs.pvalues_,4))
    top_features = pd.concat([features,f_scores,p_value], axis=1)
    top_features.columns = ["features","f_scores","p_value"]
    best_features = top_features[top_features["p_value"]<0.05]["features"]
    best_feat_df = pd.DataFrame(columns=best_features)
    best_feat_df.to_csv("best_features.csv",index=False)
    return best_feat_df.columns



def models_tuning(X_train,y_train,best_features):
    scores = []
    for model_name, mp in model_n_params_dict.items():
        mdl = RandomizedSearchCV(mp["model"],mp["params"], cv=5, return_train_score=False)
        mdl.fit(X_train[best_features],y_train)
        scores.append([model_name,mdl.best_estimator_,mdl.best_params_,mdl.best_score_])
        
#    model_evaluation(X_test,y_test,best_features,scores)
    
    return scores


def model_evaluation(X_test,y_test,best_features,scores):
    print("Entered to model evaluation")
    final_scores = []
    for i in scores:
        i[1].set_params(**i[2])
        y_pred = i[1].predict(X_test[best_features])
        mse = mean_squared_error(y_test,y_pred)
        r2s = r2_score(y_test,y_pred)
        final_scores.append([i[0],i[1],mse,r2s])
    
    max_r2s = 0
    for i in final_scores:
        temp = i[3]
        if temp > max_r2s:
            max_r2s = temp
            best_model = [i[0],i[1],max_r2s]

    filename = 'model.sav'
    pkl.dump(best_model[1], open(filename, 'wb'))
    return print("Best model is",best_model[0],best_model[2])
```

```python
df = encode_categorical_features(df)
df = remove_no_use_cols(df)
df.columns
X,y = dep_indep_seperator(df)
X_train,X_test,y_train,y_test = splits_to_train_test(X,y)
X_train,X_test = scale(X_train,X_test)
best_features = feature_selection(X_train,y_train)
scores = models_tuning(X_train,y_train,list(best_features))
model_evaluation(X_test,y_test,best_features,scores)
best_features, scores
```
