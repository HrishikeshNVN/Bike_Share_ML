# Encoding categorical variables.

```python
df["jul_sep_jun_aug"] = df["month"].apply(lambda month : 1 if month in ["July","August","September"] else 0)
df["oct_may"] = df["month"].apply(lambda month : 1 if month in ["October","May"] else 0)
df["dec_mar_nov_apr"] = df["month"].apply(lambda month : 1 if month in ["December","March","November","April"] else 0)
df["jan_feb"] = df["month"].apply(lambda month : 1 if month in ["January","February"] else 0)
```

```python
# dropping 'month', 'year' columns as it is redundant now
df.drop(['month'], axis=1, inplace=True)
df.drop(['year'], axis=1, inplace=True)
```

```python
df.head()
```

```python
def encoding_fn(df,feature):
    tdf = pd.get_dummies(df[feature], drop_first=True)
    df.drop(feature, axis=1, inplace=True)
    return pd.concat([df,tdf], axis=1)

df = encoding_fn(df,'season')
df = encoding_fn(df,'weekday')
df = encoding_fn(df,'weathersit')
```

From the above plots we observe that there is outliers in casual, windspeed and hum columns. Let's handle it.

```python
df = df[~((df.windspeed > np.nanpercentile(df.windspeed, 99)) | (df.casual > np.nanpercentile(df['casual'], 99)))]
print(df.shape)
```

```python
sns.boxplot(df["windspeed"])
```

Here we are able to remove a few outliers.

We considered 99prcentile for outlier removal. If we remove more it may affect the information we get from the whole dataset.

```python
X = df.drop(["cnt","casual","registered"], axis=1)
y = df[["cnt"]]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

Even though the numerical variables in our dataset was Minmax scaled, as we have done some outlier removal we perform Minmax scaling.

```python
nums = ['temp','hum', 'windspeed',"atemp"]
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
X_train[nums] = scalar.fit_transform(X_train[nums])
X_test[nums] = scalar.transform(X_test[nums])
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calculate_vif(X_train):
    vif_df = pd.DataFrame()
    vif_df['Features'] = X_train.columns
    vif_df['Variance Inflation Factor'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
    vif_df['Variance Inflation Factor'] = round(vif_df['Variance Inflation Factor'], 2)
    vif_df = vif_df.sort_values(by = 'Variance Inflation Factor', ascending = False)
    print(vif_df)
```

```python
X_train_lm = sm.add_constant(X_train)
lr1 = sm.OLS(y_train, X_train_lm).fit()
print(lr1.summary())
calculate_vif(X_train)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
def top_n_features(n):
    top_n_feat = []
    
    lm = LinearRegression()
    lm.fit(X_train_lm, y_train)
    rfe = RFE(lm, n)
    rfe = rfe.fit(X_train_lm, y_train)
    
    print("Top %d features selected by RFE" %n)
    ranking = list(zip(X_train_lm.columns,rfe.support_,rfe.ranking_))
    
    for i in ranking:
        if i[1]:
            top_n_feat.append(i[0])
    print(top_n_feat)
    return top_n_feat

```

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info = mutual_info_regression(X_train,y_train)
mutual_df = pd.DataFrame(mutual_info, index=X_train.columns, columns=["Importance"])
mutual_df
```

```python
mutual_df.nlargest(15,"Importance")
```

```python
top =  []
for i in range(10,22):
    top_1 = top_n_features(i)
    top.append(top_1)

print(top)
```

```python
def build_regressor(X_train,y_train,cols):
    X_train_new = X_train[cols]
    X_train_lm = sm.add_constant(X_train_new)
    regressor = sm.OLS(y_train, X_train_lm).fit()
    print(regressor.summary())
    print("=======================================================================================================")
    calculate_vif(X_train_lm)
```

```python
for i in top:
    build_regressor(X_train,y_train,i)
    print("\n\n")
```

From the RFE results, we got adjusted R2 of 84.4 with top 17 features.Also observed VIF values of those features are below 5. Which shows there is no much multicollinearity.

From the OLS results we can observe, for columns like Tuesday,jul_sep_jun_aug and holiday P value is high. We have to handle those features.

```python
top[7]
```

```python
cols = ['yr',
 'workingday',
 'atemp',
 'hum',
 'windspeed',
 'dec_mar_nov_apr',
 'jan_feb',
 'spring',
 'summer',
 'winter',
 'Monday',
 'Saturday',
 'LightSnow/LightRain',
 'Mist/Cloudy']
build_regressor(X_train,y_train,cols)
```

Finally we have selected features ['yr', 'workingday', 'atemp', 'hum', 'windspeed', 'dec_mar_nov_apr', 'jan_feb', 'spring', 'summer', 'winter', 'Monday', 'Saturday', 'LightSnow/LightRain', 'Mist/Cloudy']

Lets build final Model

```python
X_train_new = X_train[cols]
lin_reg = LinearRegression()
lin_reg.fit(X_train_new,y_train)
```

```python
y_tr_pred = lin_reg.predict(X_train_new)
sns.distplot(y_train - y_tr_pred)
```

```python
X_test_new = X_test[cols]
y_pred = lin_reg.predict(X_test_new)
```
